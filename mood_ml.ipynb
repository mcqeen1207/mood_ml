```python
import pandas as pd
import pickle
import time
```

```python
#loading trained model
start = time.time()
episode1 = 'DE' #Depressive episode
episode2 = 'ME' #Manic episode
episode3 = 'HME' #Hypomanic episode

model_depressive = pickle.load(open(''.join(['XGBoost_',episode1,'.pkl']), "rb"))
model_manic = pickle.load(open(''.join(['XGBoost_',episode2,'.pkl']), "rb"))
model_hypomanic = pickle.load(open(''.join(['XGBoost_',episode3,'.pkl']), "rb"))
end = time.time()
end-start
```

```python
#loading test dataset

data = pd.read_csv('test.csv', index_col = 0)
data
```

```python
start = time.time()
prob_de = model_depressive.predict_proba(data) # probability of depressive episode
prob_me = model_manic.predict_proba(data) # probability of manic episode
prob_hme = model_hypomanic.predict_proba(data) # probability of hypomanic episode

data_de = pd.DataFrame({'date': pd.Series(data.index), 'normal': pd.Series(prob_de[:,0]), 'episode': pd.Series(prob_de[:,1])})
data_me = pd.DataFrame({'date': pd.Series(data.index), 'normal': pd.Series(prob_me[:,0]), 'episode': pd.Series(prob_me[:,1])})
data_hme = pd.DataFrame({'date': pd.Series(data.index), 'normal': pd.Series(prob_hme[:,0]), 'episode': pd.Series(prob_hme[:,1])})

end = time.time()

end-start
```
